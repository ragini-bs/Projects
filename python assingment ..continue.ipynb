```python
def get_permutations(input_str):
    if len(input_str) == 1:
        return [input_str]

    permutations = []

    for i, char in enumerate(input_str):
        remaining_chars = input_str[:i] + input_str[i + 1:]
        partial_permutations = get_permutations(remaining_chars)

        for perm in partial_permutations:
            permutations.append(char + perm)

    return permutations

# Input the string for which you want to find permutations
input_string = input("Enter a string: ")

result = get_permutations(input_string)
print("Permutations of the string:")
for perm in result:
    print(perm)
```

```python
def is_palindrome(s):
    s = s.lower()  # Convert the string to lowercase for case-insensitive comparison
    s = s.replace(" ", "")  # Remove spaces to ignore them in the comparison
    if len(s) <= 1:
        return True
    elif s[0] != s[-1]:
        return False
    else:
        return is_palindrome(s[1:-1])

# Input the string to check for palindrome
input_string = input("Enter a string: ")

if is_palindrome(input_string):
    print(f"{input_string} is a palindrome.")
else:
    print(f"{input_string} is not a palindrome.")
```

```python
def generate_combinations(elements, k):
    if k == 0:
        return [[]]
    if not elements:
        return []

    first, rest = elements[0], elements[1:]
    without_first = generate_combinations(rest, k)
    with_first = generate_combinations(rest, k - 1)

    for combo in with_first:
        combo.append(first)

    return with_first + without_first

# Input the list of elements and the combination size
elements = input("Enter a list of elements separated by spaces: ").split()
k = int(input("Enter the combination size (k): "))

combinations = generate_combinations(elements, k)

print(f"All combinations of size {k} are:")
for combo in combinations:
    print(combo)
```

# Function Parameters and Arguments:

In Python functions, the concepts of formal parameters and actual arguments are fundamental for defining and calling functions. They refer to how values are passed into functions and how these values are handled within the function. Here's an explanation of these concepts:

Formal Parameters:

Formal parameters are placeholders or variables defined in the function header and used to receive input values when the function is called.
They represent the names and data types of the values that the function expects.
Formal parameters are used within the function's code block to perform computations or produce a result.
They are essentially variables scoped to the function, and their names are local to that function.
Example of defining a function with formal parameters:


def add_numbers(a, b):  # 'a' and 'b' are formal parameters
    result = a + b
    return result
    
    
Actual Arguments (Arguments):

Actual arguments, often referred to as arguments, are the values or expressions provided when calling a function.
They are the actual data that you pass to the function for processing.
The arguments supplied during the function call are assigned to the corresponding formal parameters within the function.
Actual arguments can be literals, variables, or expressions.
Example of calling a function with actual arguments:


num1 = 5
num2 = 7
sum_result = add_numbers(num1, num2)  # 'num1' and 'num2' are actual arguments
In the example above, a and b in the function add_numbers are formal parameters, and num1 and num2 are actual arguments passed when calling the function. Within the function, the formal parameters a and b take on the values of the actual arguments num1 and num2, and the computation is performed.

The key idea is that formal parameters define the interface and expectations of the function, and actual arguments supply the necessary data for the function to perform its task. Functions can have multiple formal parameters, and you need to provide the corresponding number of actual arguments when calling the function, ensuring that their data types match the formal parameters' data types.

```python
def greet(name="Guest", greeting="Hello"):
    """This function greets a person with a custom message."""
    print(f"{greeting}, {name}!")

# Calling the function with default arguments
greet()

# Calling the function with custom arguments
greet("Alice", "Hi")
greet("Bob")
```


In Python, you can use keyword arguments in function calls to specify which argument corresponds to which parameter, rather than relying on the order of the arguments. Keyword arguments make your code more readable and self-explanatory. Here's how you can use keyword arguments in function calls:

Provide the parameter name, followed by an equal sign (=), and then the value you want to pass.
Here's an example:

def greet(name, greeting):
    """This function greets a person with a custom message."""
    print(f"{greeting}, {name}!")

# Using keyword arguments in function calls
greet(name="Alice", greeting="Hi")
greet(greeting="Hello", name="Bob")
In this example, we have a function greet that takes two parameters, name and greeting. When calling the function, we use keyword arguments to specify which value goes with which parameter. The order of the arguments doesn't matter in this case.

When you call greet(name="Alice", greeting="Hi"), you are explicitly specifying that "Alice" should be assigned to the name parameter and "Hi" to the greeting parameter.

Similarly, when you call greet(greeting="Hello", name="Bob"), you're specifying "Hello" as the greeting and "Bob" as the name parameter.

Keyword arguments are particularly useful when a function has several parameters, and you want to make it clear which value is being assigned to each parameter, improving code readability and reducing the likelihood of errors.






```python
def calculate_sum(*args):
    """This function calculates the sum of a variable number of arguments."""
    total = 0
    for num in args:
        total += num
    return total

# Example usage of the function
result = calculate_sum(1, 2, 3, 4, 5)
print(f"The sum is: {result}")
```

# Lambda Functions and Higher-Order Functions:


Lambda functions in Python, also known as anonymous functions, are small, inline functions that are defined using the lambda keyword. They are typically used when you need a simple function for a short period of time, and you don't want to go through the process of defining a full-fledged function using the def keyword. Lambda functions have the following characteristics:

Lambda functions are commonly used in the following scenarios:

Functional Programming: Lambda functions are often used with functions like map, filter, and reduce to perform operations on sequences like lists, tuples, and dictionaries.

Short-Term or Throwaway Functions: When you need a simple function to be used temporarily for a specific task, you can define a lambda function without assigning it a name.

Simple Key Functions: Lambda functions are useful for specifying key functions for sorting or ordering data based on specific criteria.


```python
# List of tuples
data = [(1, 5), (3, 2), (2, 8), (4, 3), (5, 1)]

# Sort the list based on the second element of each tuple
sorted_data = sorted(data, key=lambda x: x[1])

# Display the sorted list
print("Sorted list based on the second element:")
print(sorted_data)
```

```python
def apply_function_to_list(numbers, func):
    """Apply a given function to each element in the list."""
    result = []
    for num in numbers:
        result.append(func(num))
    return result

# Example usage with a custom function
def square(x):
    return x ** 2

numbers = [1, 2, 3, 4, 5]
squared_numbers = apply_function_to_list(numbers, square)
print("Original list:", numbers)
print("List after applying the square function:", squared_numbers)
```

# Function Documentation and Best Practices:


Docstrings in Python are used to provide documentation and descriptions for functions, modules, classes, and methods. They serve as a way to explain the purpose, usage, and behavior of these entities. The primary purpose of docstrings is to make your code more understandable and accessible to others, including developers who might use or collaborate on your code.

Here's how to write docstrings for Python functions:

Location: Docstrings are placed immediately below the function definition, enclosed in triple-quotes (single or double) and indented consistently with the function definition.

Content:

Start with a concise, one-line summary of the function's purpose.
Include additional paragraphs to provide more detailed information, such as parameter descriptions and return value explanations.
Use clear and complete sentences.
Follow a consistent docstring style. PEP 257 (Python Enhancement Proposal) outlines a commonly accepted style guide for docstrings, including where to place sections like "Args," "Returns," and "Raises."
Sections:

Typical sections to include in a function docstring may include:
Summary: A one-line summary of the function's purpose.
Args: Descriptions of the function's parameters, their data types, and their meanings.
Returns: An explanation of the function's return value and its data type.
Raises: Any exceptions the function may raise, with explanations.
Examples: Optional examples of how to use the function.
Here's an example of a well-documented Python function with a docstring:


def calculate_sum(a, b):
    """
    Calculate the sum of two numbers.

    Args:
        a (int): The first number.
        b (int): The second number.

    Returns:
        int: The sum of the two numbers.
    """
    return a + b
The docstring in this example provides clear and detailed information about the function's purpose, parameters, and return value. It is essential to maintain a consistent and informative style for docstrings, as this documentation can be accessed using tools like help() and is valuable for developers who read or collaborate on your code.






Naming functions and variables in Python follows naming conventions and guidelines that are designed to enhance code readability and maintainability. Here are some best practices for naming functions and variables in Python:

1. Use Descriptive and Meaningful Names:

Choose names that clearly convey the purpose and functionality of the function or variable.
Favor descriptive names over short, cryptic ones. Avoid single-character variable names (e.g., x, i) except in certain cases (e.g., loop counters).
2. Follow PEP 8 Naming Conventions:

PEP 8 is the style guide for Python, and it provides conventions for naming functions and variables.
Function and variable names should be lowercase with words separated by underscores (snake_case). For example: calculate_sum, user_age.
For constants that shouldn't be changed, use uppercase with underscores (UPPER_CASE). For example: PI, MAX_VALUE.
3. Use Verb-Noun Pairing for Functions:

Choose function names that follow the pattern of "verb-noun" pairing. The verb indicates the action the function performs, and the noun describes the object or subject of the action.
Examples: calculate_sum, create_user, print_report.
4. Be Consistent:

Maintain consistency in naming within your codebase. If you use a particular naming style, stick with it throughout your project.
5. Avoid Reusing Built-in Names:

Avoid using names that conflict with built-in Python keywords or functions (e.g., str, list, sum). This can lead to unexpected behavior and errors.
6. Use Abbreviations Sparingly:

While shortening long words with abbreviations can be helpful, overusing abbreviations can make your code less readable. Only use common and well-understood abbreviations.
7. Use Underscores for Private Variables:

By convention, variables or functions intended for internal use within a module or class should have a single leading underscore. For example: _internal_data, _private_function.
8. Choose Meaningful Parameter Names:

When defining function parameters, choose parameter names that make it clear what type of data they should receive. Avoid using generic names like data or value.
9. Use Singular Nouns for Single Object Variables:

Use singular nouns for variables that represent a single object (e.g., user, car) and plurals for collections (e.g., users, cars).
10. Avoid CamelCase or PascalCase for Functions and Variables:
- Unlike some programming languages, Python prefers snake_case for naming functions and variables, as specified by PEP 8.

11. Provide Context with Module Names:
- When creating custom modules, give them meaningful names that provide context for the functions and variables they contain.

12. Comment When Necessary:
- While naming conventions aim to make code self-explanatory, it's okay to add comments for clarification, especially when the name alone might not fully explain the purpose.

13. Use Consistent Spelling and Capitalization:
- Maintain consistent spelling and capitalization throughout your code. Avoid mixing British and American English spellings.

Adhering to these naming conventions and best practices contributes to code that is easier to read, understand, and maintain. Consistency and clarity in naming functions and variables improve collaboration with other developers and help you avoid common programming pitfalls.
